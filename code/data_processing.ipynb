```python
import pandas as pd

file_path = r'C:\Users\Dell\Downloads\Data_Training\raw_data.xlsx'
df = pd.read_excel(file_path)

# Extract the first column into a list (array)

df.fillna(value=0, inplace=True)
alkaline_hydrolysis = df.iloc[:, 0].to_list()
other_electrolysis = df.iloc[:,1].to_list()
PEM = df.iloc[:,2].to_list()
SOEC = df.iloc[:,3].to_list()

print(alkaline_hydrolysis)
```

```python
import numpy as np

def normalize_list(lst):
    arr = np.array(lst)
    return (arr - arr.min()) / (arr.max() - arr.min())

# Normalize the lists
normalized_alkaline_hydrolysis = normalize_list(alkaline_hydrolysis)
normalized_other_electrolysis = normalize_list(other_electrolysis)
normalized_PEM = normalize_list(PEM)
normalized_SOEC = normalize_list(SOEC)
```

```python
data = {
    'Alkaline Hydrolysis': normalized_alkaline_hydrolysis,
    'Other Electrolysis': normalized_other_electrolysis,
    'PEM': normalized_PEM,
    'SOEC': normalized_SOEC
}

df_normalized = pd.DataFrame(data)

# Save the DataFrame to a CSV file
output_csv_path = r'C:\Users\Dell\Downloads\Data_Training\normalized_data.csv'
df_normalized.to_csv(output_csv_path, index=False)
```
